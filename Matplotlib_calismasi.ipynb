```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yas = np.array([10,20,30,30,30,40,50,60,65,70])
kilo = np.array([30,45,50,52,55,60,55,70,80,80])
```

```python
yas
```

```python
plt.plot(yas, kilo, "b")
plt.show
```

```python
plt.plot(yas, kilo, "b")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("Kilo'nun Yaşa Göre Değişimi")
plt.show
```

```python
dizi1 = np.linspace(0,10,20)
```

0 dan 10a kadar eşit aralıklarla sayı üretti

```python
dizi1
```

```python
dizi2 = dizi1 ** 3
```

```python
plt.plot(dizi1, dizi2, "g--")
```

```python
plt.plot(dizi1, dizi2, "g*-")
```

```python
plt.subplot(1,2,1)
plt.plot(dizi1,dizi2,"r*-")
plt.subplot(1,2,2) #1 sıra olacak, 2 kolon olucak, 2. grafik çizilecek
plt.plot(dizi2,dizi1,"b--")
```

```python
benimki = plt.figure()
figuraxes= benimki.add_axes([0.2,0.2,0.5,0.5])
figuraxes.plot(dizi1,dizi2,"b")
figuraxes.set_xlabel("x ekseni")
figuraxes.set_ylabel("y ekseni")
figuraxes.set_title("Başlık")
```

```python
benimki = plt.figure()
figuraxes= benimki.add_axes([0.2,0.2,0.3,0.3]) # sağdaki 2 si grafiğin boyutu ile oynuyor.
figuraxes.plot(dizi1,dizi2,"b")
figuraxes.set_xlabel("x ekseni")
figuraxes.set_ylabel("y ekseni")
figuraxes.set_title("Başlık")
```

```python
benimki2 = plt.figure()
eksen1 = benimki2.add_axes([0.1,0.1,0.8,0.8]) # soldaki 2 si grafiğin koordinatı ile oynuyor.
eksen2 = benimki2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(dizi1,dizi2,"r")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Başlık")

eksen2.plot(dizi2, dizi1, "y")
```

```python
benimki2 = plt.figure()
eksen1 = benimki2.add_axes([0.1,0.1,0.8,0.8])
eksen2 = benimki2.add_axes([1,0.5,0.3,0.3]) # soldaki 2 si grafiğin koordinatı ile oynuyor.

eksen1.plot(dizi1,dizi2,"r")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Başlık")

eksen2.plot(dizi2, dizi1, "y")
```
